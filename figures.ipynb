```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt

main_dir = os.getcwd()
```

```python
### Style Plots ###
filepath = main_dir+"/histories/history_only_style.csv"
plot_with_loss = False

history = pd.read_csv(filepath, header=0)
epochs = np.array(history["epoch"]) + 1
acc = np.array(history["accuracy"])
loss = np.array(history["loss"])
top3_acc = np.array(history["top3_acc"])
top5_acc  = np.array(history["top5_acc"])
val_acc = np.array(history["val_accuracy"])
val_loss = np.array(history["val_loss"])
val_top3_acc = np.array(history["val_top3_acc"])
val_top5_acc = np.array(history["val_top5_acc"])

plt.plot(epochs, loss, 'y', label='Training loss')
plt.plot(epochs, val_loss, 'r', label='Validation loss')
plt.title('Training and validation loss - Style')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()


if plot_with_loss:
    plt.plot(epochs, loss, 'y', label='Training loss', linestyle='dashed')
    plt.plot(epochs, val_loss, 'r', label='Validation loss', linestyle='dashed')
plt.plot(epochs, acc, 'y', label='Training acc')
plt.plot(epochs, val_acc, 'r', label='Validation acc')
plt.title('Training and validation accuracy - Style')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

if False:
    plt.plot(epochs, loss, 'y', label='Training loss', linestyle='dashed')
    plt.plot(epochs, val_loss, 'r', label='Validation loss', linestyle='dashed')
#plt.plot(epochs, acc, 'y', label='Training acc')
plt.plot(epochs, val_acc, 'r', label='Validation acc')
plt.plot(epochs, val_top3_acc, 'y', label='Validation top3-acc')
plt.plot(epochs, val_top5_acc, 'g', label='Validation top5-acc')
plt.title('Training and validation accuracy - Style')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.yticks(np.arange(0, 1, 0.25))
plt.show()
```
